OBSERVATION:

From the above plots for the years 2015-2019, we can see a decreasing trend in the happiness score for the countries in the data-sets over the years. This can be buttressed using standard deviation values derived from the describe() function. The reason for choosing standard deviation for this analysis is that it has proven to be more reliable than both mean and range values.

Since the standard deviation measures dispersion  or absolute variability of the distribution of the data relative to its mean, a low standard deviation indicates that the data points tend to be very close to the mean while a high standard deviation indicates that the data points are spread out over a large range of values. The closer the standard deviation is to zero, the closer the data points are to the mean. With a decreasing standard deviation values for the Happiness scores over the years 2015 -2019, it  can be deduced that Overall Happiness is taking a downward trend or plunging.
